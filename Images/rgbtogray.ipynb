```
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/offset_comp_772626-opt.jpg')
cv2_imshow(image)
```

```
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
import numpy as np

numpy_image = np.array(image)
gray_image = np.zeros((numpy_image.shape[0], numpy_image.shape[1]))

counter = 0
for i in range(numpy_image.shape[0]):
  for j in range(numpy_image.shape[1]):
    gray_image[i, j] = (0.299*numpy_image[i, j, 0]) + (0.587*numpy_image[i, j, 1]) + (0.114*numpy_image[i, j, 2]) 

cv2_imshow(gray_image)
print(counter)
```
