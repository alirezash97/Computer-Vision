```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


image = cv2.imread('/content/Lenna.png', cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (256, 256))
print('Original Image:')
cv2_imshow(image)
print("Histogram:")
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

```python
import numpy as np 


#Brightness
transformed_image = np.zeros_like(image)
for row in range(image.shape[0]):
  for column in range(image.shape[1]):
    # Simple linear function
    transformed_image[row, column] = (image[row, column]) + 24

print('Transformed Image:')
cv2_imshow(transformed_image)
print("Histogram:")
plt.hist(transformed_image.ravel(), 256, [0,256])
plt.show()
```

```python
#Threshold

thresholded_image = np.zeros_like(image)
for row in range(image.shape[0]):
  for column in range(image.shape[1]):
    if image[row, column] > 100:
      thresholded_image[row, column] = 255
    else:
      pass

print('Thresholded Image:')
cv2_imshow(thresholded_image)
print("Histogram:")
plt.hist(thresholded_image.ravel(), 256, [0,256])
plt.show()
```
